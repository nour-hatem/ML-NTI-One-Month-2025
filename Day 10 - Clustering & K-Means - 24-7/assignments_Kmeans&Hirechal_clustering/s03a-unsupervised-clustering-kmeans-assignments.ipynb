### Data Preprocessing

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(mrkt)
```

```python
sns.scatterplot(data=mrkt, x='Satisfaction', y='Loyalty', hue='Cluster', s=100)
plt.title('Customer Segments by Satisfaction and Loyalty')
plt.xlabel('Satisfaction')
plt.ylabel('Loyalty')
plt.legend(title='Cluster')
plt.show()
```

### Model Training & Prediction

```python
inertia = []
k_range = list(range(1, 11))
```

```python
for k in  k_range:
  model = KMeans(n_clusters=k, random_state=42)
  model.fit(scaled_data)
  inertia.append(model.inertia_)
```

### Finding the Optimal Number of Clusters (Elbow / Knee Method)

```python
print(f"len(k_range): {len(k_range)}")
print(f"len(inertia): {len(inertia)}")
```

```python
plt.figure(figsize=(8,5))
plt.xlabel('k_range')
plt.ylabel('Inertia')
plt.plot(k_range, inertia)
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
k=4
kmeans = KMeans(n_clusters=k, random_state=42)
mrkt['Cluster'] = kmeans.fit_predict(scaled_data)
```
